```python
import csv
import random as rn
import math
import operator as op
import numpy as np
#read data from csv
#with open('/home/emre/SWE546_DataMining/iris', 'rb') as csvfile:
#    lines = csv.reader(csvfile)
#    for row in lines:
#        print ', '.join(row)
        
### function splits data into training and test sets 
#randomly with standard train/test 67/33 ratio ###

import csv
import random
def loadDataset(filename, split, trainingSet=[] , testSet=[]):
    with open(filename, 'rb') as csvfile:
        lines = csv.reader(csvfile)
        dataset = list(lines)
        for x in range(len(dataset)-1):
            for y in range(4):
                dataset[x][y] = float(dataset[x][y])
            if rn.random() < split:
                trainingSet.append(dataset[x])
            else:
                testSet.append(dataset[x])                      


# test loadDataset function    
#trainingSet=[]
#testSet=[]
#loadDataset('/home/emre/SWE546_DataMining/iris', 0.67, trainingSet, testSet)
#print 'Train: ' + repr(len(trainingSet))
#print 'Test: ' + repr(len(testSet))

### find euclidian distance between two instances ###
def eDistance(instance1, instance2, length):
    distance = 0
    for x in range(length):
        distance += pow((instance1[x] - instance2[x]), 2)
    return math.sqrt(distance)

#test distance function
#array1 = [3.4, 2.7, 4.3, 'iris1']
#array2 = [2.7, 2.6, 5.2, 'iris2']
#distance = eDistance(array1, array2, 5)
#print 'Distance: ' + repr(distance)

### find k most similar neighbours from training set ###
def getNeighbours(trainingSet, testInstance, k):
    distances = []
    length = len(testInstance)-1
    for x in range(len(trainingSet)):
        dist = eDistance(testInstance, trainingSet[x], length)
        distances.append((trainingSet[x], dist))
    distances.sort(key=op.itemgetter(1))
    neighbours = []
    for x in range(k):
        neighbours.append(distances[x][0])
    return neighbours

#test getNeighbours function
#trainingSet = [[2,2,2,'a'], [4, 4, 4, 'b']]
#testInstance = [5, 5 ,5]
#k = 2
#neighbours = getNeighbours(trainingSet, testInstance, k)
#print(neighbours)

### voting for attributes and take majority vote as prediction ###
def voteForAttribute(neighbours):
    classVotes = {}
    for x in range(len(neighbours)):
        result = neighbours[x][-1]
        if result in classVotes:
            classVotes[result] += 1
        else:
            classVotes[result] = 1
    sortedVotes = sorted(classVotes.iteritems(), key = op.itemgetter(1), reverse = True)
    return sortedVotes[0][0]
#test voteForAttribute
#neighbours = [[1,2,1, 'b'], [5,2,7,'a'], [4,3,3,'b']]
#result = voteForAttribute(neighbours)
#print( ' here is the result: ' + result)

###this function calculates accuracy ### 
def calculateAccuracy (testSet, estimations):
    correct = 0
    for x in range(len(testSet)):
        if testSet[x][-1] == estimations[x]:
            correct += 1
    return (correct/float(len(testSet))) * 100.0
    
def main():
    trainingSet = []
    testSet = []
    ratio = 0.67
    loadDataset('/home/emre/SWE546_DataMining/iris',ratio, trainingSet, testSet)
    print 'Training: ' + repr(len(trainingSet))
    print 'Test: ' + repr(len(testSet))
    
    estimations = []
    k = 3
    for x in range(len(testSet)):
        neighbours = getNeighbours(trainingSet, testSet[x], k)
        result = voteForAttribute(neighbours)
        estimations.append(result)
        print('estimated: ' + repr(result) + ' actual: ' + repr(testSet[x][-1]) )
    accuracy = calculateAccuracy(testSet, estimations)
    print ('Accuracy: ' + repr(accuracy) + '%')
    
main()    
```
