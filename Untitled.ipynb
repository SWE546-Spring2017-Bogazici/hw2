```python
import numpy as np
import pandas as pd
import heapq as hp
import matplotlib.pylab as plt
import matplotlib.pyplot as plt


df = pd.read_csv(u'C:/Users/NedimAzgari/Desktop/SWE546/notes/data/iris.txt',sep=' ')
df
```

```python
A = np.hstack([
          np.matrix(df.sl).T, 
          np.matrix(df.sw).T, 
          np.matrix(df.pl).T, 
          np.matrix(df.pw).T])
c = np.matrix(df.c).T
```

```python
def Distance(x,y):
    import math 
    e = np.array(x) - np.array(y)
    e = np.sum(e**2)

    return math.sqrt(e)
```

```python
def LeaveOneOut(A, c, k):
    import heapq as hp
    N = A.shape[0]
    count = 0
    for j in range(N):
        queue = []
        for i in range(N):
            e = Distance(A[i,:] , A[j,:])
            if i != j:
                queue.append((e, i, df.iat[i,4]))
        hp.heapify(queue)
        CountList = {}
        for i in range(c.size):
            CountList[i+1] = 0 
        for i in range(k):
            nearclass = hp.heappop(queue)[2]
            CountList[nearclass] += 1;
        estimatedclass = max(CountList, key=CountList.get)
        if estimatedclass == c[j]:
            count += 1
    return (count / N) * 100
    

```

```python
B = {}
N = A.shape[0]
for i in range(N):   
    B[i] = float(LeaveOneOut(A, c, i))
    print ("for value of k %s success rate is %s" %(i, B[i])) 

```
